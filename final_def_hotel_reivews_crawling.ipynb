# 최종 코드 

```python
from selenium.webdriver import Chrome
from selenium.webdriver.common.by import By

# 다음 페이지 넘어가기
def def_next_page(browser):
    browser = browser
    # 무조건 link에 copyselector를 추가 후에 find_elements를 해야 한다.
    link = "#component_15 > div > div:nth-child(3) > div.MD > div > a.ui_button.nav.next.primary"
    next_page = browser.find_elements(By.CSS_SELECTOR,link)   

    return next_page

   # 리뷰 가져오기     
def def_crawling_review(browser):
    browser = browser
    class_reviews = []
    re_list = []
    class_reviews = browser.find_elements(By.CSS_SELECTOR,'.QewHA.H4._a > span')
    for span_review in class_reviews:
        re_list.append(span_review.text)
    return re_list
    
def def_hotel_reivews_crawling(url):
    url = url
    browser = Chrome()
    browser.get(url)

    flag = True
    n=0
    total_review = []
    next_page = None
    
    while flag:
        try:
            # 가져오 데이터를 임시변수에 저장
            return_def_crawling_review = def_crawling_review(browser)

            #가져온 데이터를 전체 데이터 리스트에 적재
            total_review.extend(return_def_crawling_review)
            #다음 페이지 이동 및 다음 페이지 여부 판별 True or False 저장
            next_page = def_next_page(browser)
            
            #  다음 페이지가 있다면 이동
            next_page[0].click()
            
        # 마지막 페이지 여부 판별 없다면 종료
        except:
            print('예외처리 발생', n)
            # 가장 오류가 많이 나는 구간이 첫번째 반복
            if  n and not next_page :
                flag = False
        n+=1
    return total_review
```

# 데이터 확인

```python
total_review = def_hotel_reivews_crawling('https://www.tripadvisor.co.kr/Hotel_Review-g294197-d306130-Reviews-Lotte_Hotel_Seoul-Seoul.html')
```
